```python
# Group means
means = df.groupby("HourVariance").mean().reset_index("HourVariance")
display(means)
```

```python
# Group Stds (equal variances ?)
stds = df.groupby("HourVariance").std(ddof=1).reset_index()
display(stds)
```

```python
# Plot groups
ax = sns.violinplot(x="HourVariance", y="deathrate", data=df)
ax = sns.swarmplot(x="HourVariance", y="deathrate", data=df,
color="white")
ax = sns.swarmplot(x="HourVariance", y="deathrate", color="black", data=means, size=10)
```

```python
import scipy.stats
import statsmodels.api as sm
#import statsmodels.stats.api as sms
import statsmodels.formula.api as smf
from statsmodels.stats.stattools import jarque_bera
```

```python

# ANOVA
lm = smf.ols('deathrate ~ HourVariance', data=df).fit()
lm
```

```python
anova_table = sm.stats.anova_lm(lm, typ=2)
display(anova_table)
```

```python
plt.figure(figsize=(25,20))
sns.set(style="darkgrid")
ax = sns.boxplot(x='HourVariance',y='deathrate', data=df)
ax.axes.set_title("Boxplots of change in death rate with increasing HourVariance parameter",fontsize=30)
ax.set_xlabel("Hour Variance",fontsize=20)
ax.set_ylabel("Death rate",fontsize=20)
ax.tick_params(labelsize=15)
plt.show()
```

```python
import seaborn as sns
sns.set()
# Create subplots 
fig, axes = plt.subplots(nrows=4, ncols=4)
fig.subplots_adjust(hspace=1)
fig.suptitle('Distributions of death rate for various HourVariance parameters')

# Generate histograms
for ax, (name, subdf) in zip(axes.flatten(), df.groupby('HourVariance')):
    sns.histplot(subdf, x='deathrate', ax=ax, kde=True)
    ax.set_title(name)

plt.show()
```

```python
sns.pairplot(df,hue='HourVariance')
```

```python
corrmat = df.corr()
hm = sns.heatmap(corrmat, cbar = True, annot = True, square = True, fmt = '.2f', annot_kws={'size':15}, 
                 yticklabels=df.columns, xticklabels=df.columns, cmap='Spectral_r')
plt.show()
```
